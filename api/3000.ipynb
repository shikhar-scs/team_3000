```python
import pandas as pd
df = pd.read_csv("OrdersList1562157417583.csv", error_bad_lines = False)
```

```python
df.dropna()
df = df[["Order#", "City", "State", "Zipcode"]]
df
```

```python
df = df.groupby("Zipcode").agg(lambda x: list(x))
df
```

```python
df["length"] = df["City"].str.len()
df = df.sort_values("length", ascending = False)
```

```python
df.to_csv("Allclusters.csv")
df
```
